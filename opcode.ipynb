```python
import gzip
def get_output_file(output_file_path, compression,compression_level):
    if compression:
        return gzip.open(output_file_path, 'wt', compresslevel=compression_level)
    else:
        return open(output_file_path, 'w')
```

```python
from capstone import Cs, CS_ARCH_X86, CS_MODE_32

def extract_opcodes_from_asm_file(file_path, output_file_path, compression=True, compression_level=9, instruction_skip=1):
    opcodes = []
    with open(file_path, 'rb') as asm_file:  # Open the file in binary mode
        # Read the content of the file
        asm_content = asm_file.read()
        # Initialize Capstone disassembler
        md = Cs(CS_ARCH_X86, CS_MODE_32)
        # Disassemble the binary content using Capstone
        with get_output_file(output_file_path, compression,compression_level) as output_file:
          for i in md.disasm(asm_content, 0x1000):
              s=''
              s= str(i.address)+' '+ str(i.mnemonic) +' '+str(i.op_str)+'\n'
              output_file.write(s)
    return opcodes

# Example usage:
file_path = 'batch 1/01IsoiSMh5gxyDYTl4CB.asm'
output_file_path = 'texts/output.txt'
opcodes = extract_opcodes_from_asm_file(file_path, output_file_path, compression=True, compression_level=9, instruction_skip=1)
#for opcode in opcodes:
 #   print(f"0x{opcode['address']:x}:\t{opcode['mnemonic']}\t{opcode['op_str']}")
print("done")
```
