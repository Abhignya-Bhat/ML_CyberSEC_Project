```python
import pandas as pd
from gensim.models import FastText
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from gensim.models import FastText
from gensim.test.utils import datapath
from gensim.models.callbacks import CallbackAny2Vec
```

```python
class EpochLogger(CallbackAny2Vec):

    '''Callback to log information about training'''


    def __init__(self):
        self.epoch = 0


    def on_epoch_begin(self, model):
        print("Epoch #{} start".format(self.epoch))


    def on_epoch_end(self, model):
        print("Epoch #{} end".format(self.epoch))
        self.epoch += 1
```

```python
model_path='saved_models/fast_text-malware.bin'
model_embed= FastText.load(model_path)
```

```python
test2 = pd.read_csv('data/dataset.csv')
test = pd.read_csv('result_embed.csv')
```

```python
test2= pd.DataFrame(test['ID']).merge(test2, on='ID').reset_index(drop = True)
test2
```

```python
test2['Opcodes'] = test2['Opcodes'].apply(lambda x: str(x).replace(',',' '))
```

```python
sample_op=test2
sample_op['Num_Opcodes'] = sample_op['Opcodes'].apply(lambda x: len(x.split()))
sample_embed=test
```

```python
merged_df = pd.merge(sample_op, sample_embed, on='ID')
merged_df=merged_df.drop('Family_x', axis=1)
merged_df = merged_df.rename(columns={'Family_y': 'Family'})
merged_df
```

```python
push_op=model_embed.wv.get_vector('push')
push_op=500*push_op
```

```python
columns_to_concat=[]
for i in range(100):
    columns_to_concat.append(str(i))
```

```python
merged_df['Vector']=merged_df[columns_to_concat].apply(lambda row: np.array(row), axis=1)
merged_df
```

```python
merged_df.drop(columns=columns_to_concat, inplace=True)
merged_df
```

```python
merged_df['Vector'] = merged_df.apply(lambda row: [val * row['Num_Opcodes'] for val in row['Vector']], axis=1)

# Print the updated DataFrame
merged_df
```

```python
merged_df['New_Vector'] = merged_df['Vector'].apply(lambda x: np.array(x) + push_op)
merged_df['New_Num_Opcodes'] = merged_df['Num_Opcodes'].apply(lambda x: int(x) + 500)
merged_df['New_Vector'] = merged_df.apply(lambda row: [x / row['New_Num_Opcodes'] for x in row['New_Vector']], axis=1)
merged_df = merged_df.drop(['Num_Opcodes', 'Vector','New_Num_Opcodes'], axis=1)
merged_df = merged_df.rename(columns={'New_Vector': 'Vector'})
merged_df
```

```python
import pandas as pd


# Expand the 'vector' column into 100 columns
expanded_vectors = pd.DataFrame(merged_df['Vector'].tolist(), columns=[i for i in range(100)])

# Concatenate the expanded vectors DataFrame with the original DataFrame
result = pd.concat([merged_df, expanded_vectors], axis=1)

# Drop the original 'vector' column if needed
result = result.drop('Vector', axis=1)

# Display the result
len(result)
```

```python
# Drop the original 'vector' column if needed
result = result.drop('Opcodes', axis=1)

# Display the result
result.head()
```

```python
result.to_csv('appended_result_embed.csv', index=False)
```
